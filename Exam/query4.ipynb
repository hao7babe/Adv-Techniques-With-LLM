
# Expected Calibration Error (ECE) Calculation

### Question
If two bins of 70%-80% and 80%-90% are used for calibration of the model, what would be the Expected Calibration Error (ECE)?

### Answer

We will calculate the **Weighted Calibration Error (WCE)** for each bin and then sum them up to get the **ECE**.

---

### Step 1: Bin Assignments

| Review ID | Confidence (%) | Bin         |
|-----------|----------------|-------------|
| 1         | 78            | 70%-80%     |
| 2         | 86            | 80%-90%     |
| 3         | 74            | 70%-80%     |
| 4         | 82            | 80%-90%     |

---

### Step 2: Accuracy and Average Confidence for Each Bin

#### For Bin 70%-80%:
- **Reviews in Bin:** Review 1, Review 3
- **Total Responses in Bin:** 2
- **Correct Responses in Bin:** 2 (both Review 1 and Review 3 are correct)
- **Accuracy:** \( \frac{2}{2} = 1.0 \)
- **Average Confidence:** \( \frac{78 + 74}{2} = 76\% \)

#### For Bin 80%-90%:
- **Reviews in Bin:** Review 2, Review 4
- **Total Responses in Bin:** 2
- **Correct Responses in Bin:** 1 (only Review 4 is correct)
- **Accuracy:** \( \frac{1}{2} = 0.5 \)
- **Average Confidence:** \( \frac{86 + 82}{2} = 84\% \)

---

### Step 3: Weighted Calibration Error for Each Bin

#### For Bin 70%-80%:
- **Weighted Calibration Error:**
  \[
  \left( \frac{2}{4} \right) \times |1.0 - 0.76| = 0.5 \times 0.24 = 0.12
  \]

#### For Bin 80%-90%:
- **Weighted Calibration Error:**
  \[
  \left( \frac{2}{4} \right) \times |0.5 - 0.84| = 0.5 \times 0.34 = 0.17
  \]

---

### Step 4: Expected Calibration Error (ECE)
\[
ECE = 0.12 + 0.17 = 0.29
\]

---

### Final Answer
The Expected Calibration Error (ECE) is **0.29**.
